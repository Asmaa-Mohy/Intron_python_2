### 1. Lists
is a collection which is ordered and mutable(changeable). Allows duplicate members. is homogeneous/heterogenous elements (int, float, string, etc.)

##### Creation
`list_name` = [`items`]
OR
`list_name` = list((`items`))


```python
#created  list

list_ = ['hello', 1, 10.5,'pytho',6.8]
list_
```

```python
# length of the list_

L = len(list_)
L
```

##### Access items
`list_name` [`i`] where `i` is index of element

```python
# first element
list_[0]
```

```python
list_[-1] # -1 implies last element of the list
```

```python
#Slicing To get a specific set of elements from the list

list_[:4]  # left-closed type data retrieval
```

```python
list_[1:4]  # 1-3 are retrived
```

```python
list_[2:]  # 2-4 elements are retrieved. Leaving right-end empty implies last element in the list
```

```python
list_[:-1] # only the last element is skipped
```

```python
list_[:-2] # last and last-before elements are skipped (-1 and -2 respectively)
```

```python
list_[::-1]  # to reverse the order
```

##### Delete

```python
del list_[4]

list_ # the list_[4] was deleted and replaced by the next element in the list
# there are anther methods to removes the specified item

#list_.remove(item)
#list_.pop(index)
#list_.clear()
```

##### Add

```python
# Add item in  to the end of the list Using the append()

list_.append(6.8)
list_
```

```python
#Add an item at the specified index using the insert() method

list_.insert(3,'data')
list_
```

### Sets
A set is an unordered ,Immutable and dynamic collection with . Set is another data type that we often use in data wrangling. It is used to eliminate duplicate values from a list .


##### Creation
`set_name` = {`items`}
OR
`set_name` =set([`items`])


```python
set_ = set(list_)
set_
```

###### Add

```python
set_.add(False)  # values can be added - dynamic 
#Note : cannot add 'True' 
#It returns only the unique values from the list_ in no specific order.
set_
```

```python
#To add more than one item to a set use the update() method
set_.update([8,9])
set_
```

### Dictionaries
A dictionary consists of key-value pairs in no specific order. The key-value pairs are stored in an associative array. Each key maps to a value or a list or another dictionary.is homogeneous/heterogenous (key,value)


##### Creation

- Method 1:

  `dictionary _name` =
        
        {`key`:`Value`,
        `k2`:`V2`,
        .....
        ....
        }
- Method 2:
   
   dict([ `key`:`Value`,.....])
                               
- Method 3:                            
   
   dict(`key`=`Value`, .....)
   
- Method 4 : list comprehension technique

   [`key`:`Value` for`key` in sequence]


```python
# Lets store age of people as dictionary
dict_ = {"Amr": 32, "Mona": 23, "Hand": 12, "Alaa": 15, "Sara": 18}
dict_
```

```python
# method 2
dict([('Amr', 32), ('Mona', 23), ('Hand', 12)])
```

```python
# method 3
dict(Amr=32, Mona=23, Hand=12)
```

```python
# method 4 
{x: 2*x for x in range(5)}
```

###### Access

```python
# To retrieve values
dict_['Amr']
```

```python
#by get() method
dict_.get('Amr')
```

```python
# To get list of keys
dict_.keys()
```

```python
# To get list of values
dict_.values()
```

###### print a dictionary

```python
for k, v in dict_.items():print(k, v)
```

###### reverse the key-value pairs

```python
dict_reversed = {v:k for k, v in dict_.items()}  
dict_reversed
```
