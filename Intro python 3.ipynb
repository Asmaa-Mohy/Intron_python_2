###### Types of Arguments 

- ###### Positional Arguments:
values are copied to their corresponding parameters in order. we need to pass arguments in the order in which they are defined.


```python
def func(name, job):
    print(name, 'is a', job)

func('Ramy', 'developer')
func('developer', 'Bob')
```


- ###### Keyword Arguments:
To avoid positional argument confusion, you can pass arguments using the names of their corresponding parameters.It is possible to combine positional and keyword arguments in a single call. If you do so, specify the positional arguments before keyword arguments.

```python
def func(name, job):
    print(name, 'is a', job)

func(name='Bob', job='developer')
func(job='developer', name='Bob')
func('Bob',job='developer') 
#func('developer', name='Bob') #eror
#func(name='Bob','developer') #error
```


- ###### Default Arguments:
we can specify default values for arguments when defining a function. The default value is used if the function is called without a corresponding argument.

```python
def func(name, job='developer'):
    print(name, 'is a', job)

func('Bob', 'manager')
func('Bob')
```


- ###### Variable Length Arguments (*args)
var-args are useful when you want to create functions that take unlimited number of arguments. it collects all the unmatched arguments into a tuple.

```python
def my_min(*args):
    result = args[0]
    for num in args:
        if num < result:
            result = num
    return result

my_min(4, 5, 6, 7, 2)
```


- ###### Variable Length Arguments (**kwargs)
it only works for keyword arguments.It collects them into a new dictionary, where the argument names are the keys, and their values are the corresponding dictionary values.

```python
def kwarg_type_test(**kwargs):
    print(kwargs)

kwarg_type_test(a="hi")
kwarg_type_test(roses="red", violets="blue")
```

### Class & Object :

is the code templates to create objects.
- `__init__` is simply used to initialize several attributes of a class.
- `self` is the first argument, to access the attributes of the class. 

```python
from IPython.display import Image

review = "This was a horrible, terrible movie."

Image(filename='python oop_1.png')
```

```python
class car:
    def __init__(self, year, mpg,speed):
        # attributes
        self.year = year     # car model's year
        self.mpg =  mpg       # mileage
        self.speed =speed     # current speed
    # methods
    def accelerate(self):
        return self.speed + 20

    def brake(self):
        return self.speed - 50
 
```
